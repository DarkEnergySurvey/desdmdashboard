# How to load desdmdashboard data and make custom plots

Import the functions that allow to load data from the desdmdashboard database directly into a pandas DataFrame.

These functions require you to have a [desdmdashboard] section in your .desservices file similar to


    [desdmdashboard]
    user = my_username
    passwd = my_password
    # web api url of the desdmdashboard
    api_url = http://desdash.cosmology.illinois.edu/dev/desdmdashboard/monitor/api

```python
from desdmdashboard_remote.receivedata import to_pandas
```

## Load a single metric

The owner keyword is optional. In case it is not specified, the function assumes that the metric specified in the first argument by its name, can be found in the namespace of the user declared in the .desservices.ini file.

```python
df = to_pandas.get_metric_dataframe('desar_conn_to_any', owner='gdaues')
```

```python
df.describe()
```

```python
df.tail()
```

### Plotting
Load the pylab inline environment to plot in this notebook inline.
Increase the standard figure size.

```python
pylab inline
```

```python
figsize(10, 6)
```

```python
fig = df.last('7D').plot(style='.-', colormap='jet', linewidth=2)
```

## Load multiple metrics
We need to resample the datapoints to a common frequency for all metrics.

```python
df2 = to_pandas.get_multimetric_dataframe((('gdaues', 'dessci_read_GB',),
                                          ('gdaues', 'dessci_write_GB',),
                                          ('gdaues', 'dessci_mydb_GB'),), 
                                          resample='D')
```

```python
ax = df2.plot(style='*-', colormap='jet', linewidth=3)
fig = ax.get_figure()
fig.tight_layout()
title('DESSCI Summary')
```

```python
fig.savefig('dessci_summary.png')
```
